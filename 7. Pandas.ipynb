## Types of joining the Dataframes

```python
df1 = pd.DataFrame({'Num':np.arange(65,71), 'Char':[chr(i) for i in range(65,71)]})
df1
```

```python
df2 = pd.DataFrame({'Num':np.arange(71,76), 'Char':[chr(i) for i in range(71,76)]})
df2
```

## 1. Append()

```python
display(df1,df2)
```

```python
df1.append(df2)
```

```python
df1.append(df2, ignore_index=1)
```

```python
df3 = pd.DataFrame({'Num_new':np.arange(71,76),
                    'Char_new':[chr(i) for i in range(71,76)]})
df3
```

```python
display(df1,df3)
```

```python
df1.append(df3,ignore_index=1)
```

```python
df1.append(df3,axis = 1, ignore_index=1)
```
